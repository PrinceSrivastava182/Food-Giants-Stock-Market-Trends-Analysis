```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("C:\\Users\\Welcome\\Desktop\\FOOD GIANTS\\Datasets\\Food Giants.csv")
```

```python
data
```

```python
print("Dataset Information:")
data.info()
print("\nSummary Statistics:")
print(data.describe())
```

```python
print("\nMissing Values:")
print(data.isnull().sum())
```

```python
duplicates = data.duplicated().sum()
print(f"\nDuplicate Rows: {duplicates}")
```

```python
numeric_cols = data.select_dtypes(include=['number']).columns
for col in numeric_cols:
    plt.figure(figsize=(10, 5))
    sns.histplot(data[col], kde=True, bins=30, color='royalblue', edgecolor='black')
    plt.title(f'Distribution of {col}', fontsize=14, fontweight='bold')
    plt.xlabel(col, fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
```

```python
# Correlation heatmap (only numeric columns)
plt.figure(figsize=(12, 8))
numeric_data = data.select_dtypes(include=['number'])
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Correlation Heatmap", fontsize=14, fontweight='bold')
plt.show()
```

```python
# Box plots for outlier detection
for col in numeric_cols:
    plt.figure(figsize=(10, 5))
    sns.boxplot(x=data[col], color='teal')
    plt.title(f'Box Plot of {col}', fontsize=14, fontweight='bold')
    plt.grid(axis='x', linestyle='--', alpha=0.7)
    plt.show()
```

```python
# Pairplot to visualize relationships between numerical variables
sns.pairplot(data[numeric_cols], diag_kind='kde', corner=True)
plt.show()
```

```python
# Violin plots to show distribution of numerical variables
for col in numeric_cols:
    plt.figure(figsize=(10, 5))
    sns.violinplot(x=data[col], palette='coolwarm')
    plt.title(f'Violin Plot of {col}', fontsize=14, fontweight='bold')
    plt.grid(axis='x', linestyle='--', alpha=0.7)
    plt.show()
```

```python
# Scatter plot to visualize relationships between the two most correlated variables
if len(numeric_cols) >= 2:
    top_corr = numeric_data.corr().abs().unstack().sort_values(ascending=False)
    top_corr = top_corr[top_corr < 1].reset_index().iloc[0]
    plt.figure(figsize=(8, 5))
    sns.scatterplot(x=data[top_corr['level_0']], y=data[top_corr['level_1']], alpha=0.7, color='purple')
    plt.title(f'Scatter Plot of {top_corr["level_0"]} vs {top_corr["level_1"]}', fontsize=14, fontweight='bold')
    plt.xlabel(top_corr['level_0'], fontsize=12)
    plt.ylabel(top_corr['level_1'], fontsize=12)
    plt.grid(True)
    plt.show()
```

```python
# Count plots for categorical variables
categorical_cols = data.select_dtypes(include=['object']).columns
for col in categorical_cols:
    plt.figure(figsize=(12, 5))
    sns.countplot(y=data[col], order=data[col].value_counts().index, palette='viridis')
    plt.title(f'Count Plot of {col}', fontsize=14, fontweight='bold')
    plt.xlabel('Count', fontsize=12)
    plt.ylabel(col, fontsize=12)
    plt.show()
```
